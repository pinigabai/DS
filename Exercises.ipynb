```python
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def greet(self):
        print(f'Hello {self.name}! Your age is {self.age}.')
person = Person('Pini',38)
person.greet()

class Rectangle:
    def __init__(self,length = 1.0,width = 1.0):
        self.length = length
        self.width = width
    def area(self):
        print(f'Area: {self.length*self.width}')
    def perimeter(self):
        print(f'Perimeter: {2*self.length+2*self.width}')
rectangle = Rectangle()
rectangle.area()
rectangle.perimeter()

class BankAccount:
    def __init__(self,balance):
        self.__balance = balance
    def deposit(self,amount):
        self.__balance += amount
    def withdraw(self,amount):
        self.__balance -= amount
    def get_balance(self):
        return self.__balance
account = BankAccount(100)
account.deposit(100)
account.withdraw(50)
account.get_balance()

class Vehicle:
    def __init__(self,brand,model):
        self.brand = brand
        self.model = model
    def describe(self):
        print(f'{self.brand} | {self.model}')
class Car(Vehicle):
    def __init__(self,brand,model,doors):
        super().__init__(brand,model)
        self.doors = doors
    def describe(self):
        print(f'{self.brand} | {self.model} | {self.doors} doors')
vehicle = Vehicle('Hyundai','i20')
vehicle.describe()
car = Car('Hyundai','i20',4)
car.describe()

class MathOperations:
    @staticmethod
    def add(a,b):
        return a+b
    @classmethod
    def is_even(cls,num):
        if num%2 == 0:
            return True
        return False
print(MathOperations.add(2,4))
print(MathOperations.is_even(4))

class Dog:
    def speak(self):
        print('woof')
class Cat:
    def speak(self):
        print('meow')
class Bird:
    def speak(self):
        print('chirp')
def animal_sound(animal):
    animal.speak()
dog = Dog()
animal_sound(dog)
cat = Cat()
animal_sound(cat)
bird = Bird()
animal_sound(bird)

class Temperature:
    def __init__(self,celsius):
        self.__celsius = celsius
    @property
    def celsius(self):
        return self.__celsius
    @celsius.setter
    def celsius(self,value):
        self.__celsius = value
    @property
    def fahrenheit(self):
        return self.__celsius*9/5+32
    @fahrenheit.setter
    def fahrenheit(self,value):
        self.__celsius = (value-32)*5/9
temp = Temperature(25)
print(f'Celsius: {temp.celsius}째C')
print(f'Fahrenheit: {temp.fahrenheit}째F')
temp.fahrenheit = 77
print(f'Celsius after setting Fahrenheit to 77째F: {temp.celsius}째C')

from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
class Circle(Shape):
    pi = 3.14
    def __init__(self,radius):
        self.radius = radius
    def area(self):
        print(f'Circle area: {Circle.pi*self.radius**2}')
class Square(Shape):
    def __init__(self,side):
        self.side = side
    def area(self):
        print(f'Square area: {self.side*self.side}')
circle = Circle(2)
circle.area()
square = Square(2)
square.area()

class Vector:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def __add__(self,other):
        return Vector(self.x+other.x,self.y+other.y)
    def __sub__(self,other):
        return Vector(self.x-other.x,self.y-other.y)
    def __str__(self):
        return f'Vector({self.x},{self.y})'
v1 = Vector(2,3)
v2 = Vector(4,1)
v3 = v1+v2
print(f'Addition: {v3}')
v4 = v1-v2
print(f'Subtraction: {v4}')

class Book:
    def __init__(self,title,author,year):
        self.title = title
        self.author = author
        self.year = year
class Library:
    books = []
    @classmethod
    def add_book(cls,book):
        cls.books.append({'title':book.title,
                          'author':book.author,
                          'year':book.year})
    @classmethod
    def list_books(cls):
        for book in cls.books:
            print(f'Title: {book['title']}, Author: {book['author']}, Year: {book['year']}')
book1 = Book('The Odyssey','Homer',1996)
book2 = Book('Harry Potter and the Goblet of Fire','J.K. Rowling',2000)
book3 = Book('The Road','Cormac McCarthy',2006)
Library.add_book(book1)
Library.add_book(book2)
Library.add_book(book3)
Library.list_books()
```

```python
num = int(input('Enter the number of people:'))
prob = 1
days = 364
for _ in range(num-1):
    prob *= days/365
    days -= 1
print(f'The probability that at least two people share the same birthday is {(1-prob)*100:.2f}%')
```
